```python
import numpy as np
import matplotlib.pyplot as plt

xs, ys = np.meshgrid(np.arange(0, 10), np.arange(0, 10))

def make_z_matrix(tl, tr, bl, br):
    z_matrix = np.zeros((10, 10))

    for x in range(0, 10):
        for y in range(0, 10):
            xdiff = ((tl-tr)/10)*x
            ydiff = ((tl-bl)/10)*y

            z_matrix[x, y] = tl+xdiff+ydiff

    return z_matrix

def make_z_matrix_kron(tl, tr, bl, br):
  return np.kron(np.array([
    [tl, tr],
    [bl, br],
  ]), np.ones((5, 5)))

zs = make_z_matrix(0, 10, 20, 0)

ax = plt.axes(projection='3d')
colormap = plt.get_cmap('plasma')

surf = ax.plot_surface(xs, ys, zs, cmap=colormap)

plt.colorbar(surf, ax=ax, shrink=0.5, aspect=5)

ax.set_title("3D Heatmap Kjøleskap")
ax.set_xlabel('x bredde')
ax.set_ylabel('y høyde')
ax.set_zlabel('z temp.')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

matrix_size = 50

def upscale_matrix(matrix, n=10):
    (tl, tr), (bl, br) = matrix
    
    new_matrix = np.zeros((n, n))
    
    dly = (bl-tl)/n
    dry = (br-tr)/n

    for y in range(n):
        for x in range(n):
            l = tl + dly*y
            r = tr + dry*y
            dbx = (r-l)/n
            z = l + dbx*x
            new_matrix[y, x] = z
            
    return new_matrix

zs = upscale_matrix(np.array([
    [0, 5],
    [5, 0],
]), matrix_size)

xs, ys = np.meshgrid(np.arange(matrix_size), np.arange(matrix_size, 0, -1))

ax = plt.axes(projection='3d')
colormap = plt.get_cmap('plasma')

surf = ax.plot_surface(xs, ys, zs, cmap=colormap)

plt.colorbar(surf, ax=ax, shrink=0.5, aspect=5)

ax.set_title("3D Heatmap Kjøleskap")
ax.set_xlabel('x bredde')
ax.set_ylabel('y høyde')
ax.set_zlabel('z temp.')

plt.show()
```
